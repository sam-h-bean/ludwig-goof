```python
import os
for name in ['training', 'testing']:
    with open('drive/MyDrive/Colab Notebooks/mnist/mnist_dataset_{}.csv'.format(name), 'w') as output_file:
        print('=== creating {} dataset ==='.format(name))
        output_file.write('image_path,label\n')
        for i in range(10):
            path = 'drive/MyDrive/Colab Notebooks/mnist/mnist_png/{}/{}'.format(name, i)
            for file in os.listdir(path):
                if file.endswith(".png"):
                    output_file.write('{},{}\n'.format(os.path.join(path[36:], file), str(i)))
```
